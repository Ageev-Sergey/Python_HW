```python
# Задача 40: Работать с файлом california_housing_train.csv, который находится в папке sample_data.
# Определить среднюю стоимость дома, где кол-во людей от 0 до 500 (population)
```

```python
import pandas
df = pandas.read_csv('C:/gb/Python_second_block/CW-HW_11/california_housing_train.csv')
```

```python
df[df['population'] <= 500]
```

```python
sum = 0
col = 0
print(df[['median_house_value']][df['population']<= 500])
for i in df['median_house_value'][df['population']<= 500]:
    sum += i
    col +=1
print(f'средняя стоимость дома, где кол-во людей от 0 до 500 = {round(sum/col, 3)}')
```
